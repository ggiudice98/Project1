```python
import pandas as pd
```

```python
#Import flight data csv
flightdata_file="Resources/CompleteData.csv"
```

```python
#Read csv file with Pandas
flightdata_df=pd.read_csv(flightdata_file)

flightdata_df.head(100)
```

```python
na_df = flightdata_df.isna().sum()
na_df
```

```python
#Working Project df with valuable data
Project_flightdata_df=flightdata_df[['FL_DATE','OP_UNIQUE_CARRIER','ORIGIN', 'DEST','DEP_TIME', 'CRS_DEP_TIME', 'TAXI_OUT', 'DEP_DELAY', 'AIR_TIME',
       'DISTANCE', 'CANCELLED','TAIL_NUM','ICAO TYPE', 'YEAR OF MANUFACTURE','TEMPERATURE', 'CLOUD_COVER', 'ACTIVE_WEATHER']]

Project_flightdata_df.head()
```

```python
#Renaming Columns
Project_flightdata_df= Project_flightdata_df.rename(columns={'FL_DATE': 'Flight Date', 'OP_UNIQUE_CARRIER': 'Carrier','DEST':'Destination','DEP_TIME':'Departure Time','CRS_DEP_TIME':'CRS Departure Time','TAXI_OUT':'Taxi Out','DEP_DELAY':'Departure Delay','AIR_TIME':'Air Time','DISTANCE':'Distance','CANCELLED':'Cancelled','TAIL_NUM':'Tail Number','ICAO TYPE':'ICAO Type', 'YEAR OF MANUFACTURE':'Manufacture Year','TEMPERATURE':'Temperature','CLOUD_COVER':'Cloud Cover','ACTIVE_WEATHER':'Active Weather'})
```

```python
na_df2=Project_flightdata_df.isna().sum()
na_df2
```

```python
#Cleaned Flight Dataset
Clean_flightdata_df=Project_flightdata_df.dropna()
Clean_flightdata_df
```
